```python
try:
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    driver.delete_all_cookies()
    driver.get("https://moviesmod.info/page/2/")
    time.sleep(3)
    print(driver.title)
    # source = driver.page_source.encode("utf-8")
    # print(source)
    element = driver.find_element(By.CLASS_NAME, 'post-cards')
    print(type(element))
    print(element)
    elements = element.find_elements(By.XPATH, './/*')
    elements_id = []
    elements_title = []
    elements_links = []
    print(len(elements))
    for e in elements:
      print(e.get_attribute('href'))
      if e.get_attribute('id').startswith('post'):
        elements_id.append(e)

      if e.get_attribute('title'):
        elements_title.append(e.get_attribute('title'))

      if e.get_attribute('href'):
        elements_links.append(e.get_attribute('href'))

    elements_title = list(set(elements_title))
    elements_links = list(set(elements_links))

    print(len(elements_id))
    print(len(elements_title))
    print(len(elements_links))

    # for element in elements:
    #   id.append(element.id())
    #   print(id)
    # xpath = '//*[@id="post-'+ str(id) +'"]/a'
    # print(xpath)
    # elements = driver.find_elements(By.XPATH, xpath)
    driver.close()
except Exception as e:
    print(e)
```
